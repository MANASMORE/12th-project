```python
def add():
    while True:
        ask=int(input('Do you want to enter multiple pokemons?: \n'))
        if ask==1:
            ask2=int(input('How many pokemons do you want to enter?: \n'))
            for i in range(0,ask2):
                n=str(input('Enter the name of pokemon: \n'))
                t1=str(input("Enter Primary Type of Pokemon: \n"))
                while t1 not in df.type1.unique():
                    print('Please Enter correct Primary T')
                    t1=str(input("Enter Primary Type of Pokemon: \n"))
                t2=str(input("Enter Secondary Type of Pokemon: \n"))
                while t2 not in ['poison', 'flying', 'dark', 'electric', 'ice', 'ground','fairy', 'grass', 'fighting', 'psychic', 'steel', 'fire', 'rock','water', 'dragon', 'ghost', 'bug', 'normal','none']:
                    print('Please Enter correct Secondary T')
                    t2=str(input("Enter Primary Type of Pokemon: \n"))
                if t2=='none':
                    t2=np.NaN
                gen=int(input('Enter the Generation of your Pokemon: \n'))
                h=int(input('Height: \n'))
                w=int(input('Weight: \n'))
                l=str(input('Legendary: \n'))
                p_no=len(df.index)+1
                df.loc[len(df.index)]=[p_no,n,t1.lower(),t2,gen,h,w,l]
        elif ask==2:
            n=str(input('Enter the name of pokemon: \n'))
            t1=str(input("Enter Primary Type of Pokemon: \n"))
            while t1 not in df.type1.unique():
                print('Please Enter correct Primary T')
                t1=str(input("Enter Primary Type of Pokemon: \n"))
            t2=str(input("Enter Secondary Type of Pokemon: \n"))
            while t2 not in ['poison', 'flying', 'dark', 'electric', 'ice', 'ground','fairy', 'grass', 'fighting', 'psychic', 'steel', 'fire', 'rock','water', 'dragon', 'ghost', 'bug', 'normal','none']:
                print('Please Enter correct Secondary T')
                t2=str(input("Enter Primary Type of Pokemon: \n"))
            if t2=='none':
                t2=np.NaN
            gen=int(input('Enter the Generation of your Pokemon: \n'))
            h=int(input('Height: \n'))
            w=int(input('Weight: \n'))
            l=str(input('Legendary: \n'))
            p_no=len(df.index)+1
            df.loc[len(df.index)]=[p_no,n,t1.lower(),t2,gen,h,w,l]
        else:
            print('Please enter valid input')
add()
```

```python
def search():
    query1=int(input('Search p:'))
    if query1==1:
        pno=int(input('Enter Pokedex No:'))
        print(df.loc[[pno-1]])
    elif query1==2:
        nam=str(input('Enter Pokemon Name:'))
        print(df.loc[df['name']==nam])
```

```python
def remove():
    query2=int(input('Remove:'))
    if query2==1:
        pno1=int(input('Enter Pokedex No:'))
        a=df.drop([pno1-1],inplace=True)
        print(df)
    elif query2==2:
        nam1=str(input("Enter Pokemon Name:"))
        b=df.drop(df.index[df['name']==nam1],inplace=True)
        print(df)
```

```python
while True:
    query=int(input('What do you want to do?\n Press 1 for adding a Pokemon.\n Press 2 for searching a Pokemon.\n Press 3 for removing a Pokemon:'))
    if query==1:
        add()
    elif query==2:
        search()
    elif query==3:
        remove()
    else:
        print('Please Enter a Valid Input')
```
