```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('pokemon.csv')
df
```

```python
df.columns
```

```python
df1=df[['name','pokedex_number','type1','type2','generation','height_m','weight_kg','is_legendary']]
df1=df1.set_index('pokedex_number')
df1
df1.to_csv('PokedexFinal.csv')
```

```python
offset=.4
a=plt.hist(df1['type1'],bins=18,ec='k',align='mid')
plt.xlabel('Primary type')
plt.ylabel('No.of Pokemons')
plt.xticks(rotation=90)
plt.xticks(a[1][0:] + offset)
plt.show()
```

```python
b=plt.hist(df1['type2'].dropna(),bins=18,ec='k',align='mid')
plt.xlabel('Secondary type')
plt.ylabel('No.of Pokemons')
plt.xticks(rotation=90)
plt.xticks(b[1][0:] + offset)
plt.show()
```

```python
bins1=np.arange(3)-0.5
c=plt.hist(df1['is_legendary'],bins1,ec='k')
plt.xlabel('Is the pokemon Legendary')
plt.ylabel('No. of pokemons')
plt.xticks(df1['is_legendary'])
plt.show()
```

```python
bins=np.arange(10)-0.25
d=plt.hist(df1['generation'],bins,width=0.5,align='mid')
plt.xlabel('Generation Number')
plt.ylabel('No. of pokemons')
plt.xticks(np.arange(8))
plt.xlim([0, 8])
plt.show()
```
